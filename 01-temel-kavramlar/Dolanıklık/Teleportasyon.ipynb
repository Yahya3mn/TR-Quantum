# Teleportasyon (Kuantum ışınlama)
Kuantum ışınlama, veya kısaca ışınlama, bir göndericinin (Alice) iki bitlik klasik iletişim ile birlikte paylaşılan bir kuantum dolanıklık durumu (spesifik olarak bir e-bit) kullanarak bir alıcıya (Bob) bir kuantum bitini (kubit) iletmesi protokolüdür.
*Işınlanma* adı, maddenin fütüristik bir süreçle bir yerden başka bir yere taşındığı bilim kurgu kavramını çağrıştırmak için kullanılmıştır, ancak kuantum ışınlanmada maddenin ışınlanmadığı, aslında ışınlanan şeyin kuantum bilgisi olduğu anlaşılmalıdır.
Işınlanma için kurulum şu şekildedir.
Alice ve Bob'un bir e-bit paylaştığını varsayalım: Alice bir kuantum biti $\mathsf{A}$'ya sahiptir, Bob bir kuantum biti $\mathsf{B}$'ye sahiptir ve bu ikili $(\mathsf{A},\mathsf{B})$ birlikte $\vert\phi^+\rangle$ durumunda bulunur.
Örneğin, Alice ve Bob geçmişte aynı yerdeydiler, $\mathsf{A}$ ve $\mathsf{B}$ kuantum bitlerini $\vert \phi^+ \rangle$ durumunda hazırladılar ve sonra her biri kendi yoluna kuantum bitini elinde götürdü.
Ya da, bu paylaşılan e-bit'i oluşturmak için üçüncü bir tarafın dahil olduğu veya karmaşık bir dağıtık süreç gibi farklı bir süreç kullanılmış olabilir.
Bu ayrıntılar, teleportasyon protokolünün bir parçası değildir.
Alice daha sonra Bob'a iletmek istediği üçüncü bir kuantum bit $\mathsf{Q}$'ya sahip olur.
Kuantum bit $\mathsf{Q}$'nun durumu Alice ve Bob için *bilinmeyen* olarak kabul edilir ve bu konuda hiçbir varsayımda bulunulmaz.
Örneğin, kuantum bit $\mathsf{Q}$ Alice ve Bob'un erişemeyeceği bir veya daha fazla sistemle dolanık olabilir.
Alice'in kuantum bitini $\mathsf{Q}$'yu Bob'a iletmek istediğini söylemek, Alice'in Bob'un, protokolün başlangıcında $\mathsf{Q}$'nun bulunduğu durumla aynı durumda olan ve $\mathsf{Q}$'nun diğer sistemlerle sahip olduğu tüm korelasyonlara sahip bir kuantum bitine sahip olmasını istediği anlamına gelir, sanki Alice $\mathsf{Q}$'yu Bob'a fiziksel olarak vermiş gibi.
Alice'in kuantum bitini $\mathsf{Q}$'yu Bob'a fiziksel olarak gönderdiğini ve bu kuantum bitinin Bob'a ulaşırken değiştirilmeden veya bozulmadan ulaştığını varsayabiliriz. Bu durumda Alice ve Bob'un görevi tamamlanmış olur.
Ancak, ışınlanma bağlamında bunun mümkün olmadığını varsayıyoruz; Alice kuantum bitlerini Bob'a doğrudan gönderemez.
Ancak, Bob'a klasik bilgi gönderebilir.
Bu varsayımlar çeşitli durumlarda makul varsayımlardır.
Örneğin, Alice Bob'un tam yerini bilmiyorsa veya aralarındaki mesafe çok büyükse, bugünün teknolojisiyle veya öngörülebilir gelecekte bir kuantum bitini fiziksel olarak göndermek en azından zor olacaktır.
Ancak, günlük deneyimlerimizden bildiğimiz gibi, bu koşullar altında klasik bilgi iletimi oldukça basittir.
Bu noktada, Alice ve Bob'un ortak bir e-bit kullanmaya bile gerek kalmadan görevlerini yerine getirip getiremeyecekleri sorusu akla gelebilir.
Başka bir deyişle, klasik iletişim kullanarak bir kuantum bitini iletmenin bir yolu var mı?
Cevap hayır, klasik iletişim kullanarak kuantum bilgisini iletmek mümkün değildir.
Bu, temel kuantum bilgi teorisi kullanılarak matematiksel olarak kanıtlamak çok zor değildir, ancak alternatif olarak, klonlama teoremini düşünerek, yalnızca klasik iletişim kullanarak kuantum bitleri iletmenin imkansız olduğunu da kanıtlayabiliriz.
Klasik iletişim kullanarak kuantum bilgisi göndermenin bir yolu olduğunu düşünün.
Klasik bilgi kolayca kopyalanabilir ve yayınlanabilir, bu da Alice'den Bob'a yapılan herhangi bir klasik iletimin ikinci bir alıcı (örneğin Charlie) tarafından da alınabileceği anlamına gelir.
Ancak Charlie, Bob'un aldığı klasik iletişimin aynısını alırsa, o da kuantum bitinin bir kopyasını elde edemez mi?
Bu, $\mathsf{Q}$'nun klonlandığını gösterir, ancak klonlanmama teoremi ile bunun imkansız olduğunu zaten biliyoruz. Dolayısıyla, klasik iletişim kullanarak kuantum bilgisi göndermenin bir yolu olmadığı sonucuna varırız.
Ancak Alice ve Bob'un bir e-bit paylaştığı varsayımı geçerli olduğunda, Alice ve Bob görevlerini yerine getirebilirler.
Kuantum ışınlama protokolü tam olarak bunu yapar.

## Protokol
İşte teleportasyon protokolünü açıklayan bir kuantum devre şeması:



Şema, Alice ve Bob arasındaki ayrımı iki diyagonal tel ile temsil ederek biraz stilize edilmiştir. Bu teller, Alice'den Bob'a gönderilen klasik bitleri temsil eder. Bunun dışında şema, sıradan bir kuantum devre şemasıdır.
Kubit isimleri, başlangıç durumlarının da gösterilebilmesi için sol tarafta değil, tellerin üzerinde gösterilmiştir (uygun olduğunda genellikle bu şekilde yapacağız).
Ayrıca, $X$ ve $Z$ kapılarının *klasik* kontrolleri olduğu da unutulmamalıdır. Bu, kapıların, bu klasik kontrol bitlerinin sırasıyla $0$ veya $1$ olmasına bağlı olarak uygulanmadığı veya uygulandığı anlamına gelir.
Sözlerle ifade etmek gerekirse, ışınlanma protokolü şu şekildedir:
1.  Alice, $\mathsf{Q}$ kontrol ve $\mathsf{A}$ hedef olmak üzere $(\mathsf{A},\mathsf{Q})$ çifti üzerinde kontrollü NOT işlemi gerçekleştirir ve ardından $\mathsf{Q}$ üzerinde Hadamard işlemi gerçekleştirir.
2.  Alice daha sonra her iki durumda da standart temel ölçüme göre hem $\mathsf{A}$ hem de $\mathsf{Q}$'yu ölçer ve klasik sonuçları Bob'a iletir. $\mathsf{A}$ ölçümünün sonucuna $a$, $\mathsf{Q}$ ölçümünün sonucuna $b$ diyelim.
3.  Bob, Alice'den $a$ ve $b$ değerlerini alır ve bu bitlerin değerlerine bağlı olarak şu işlemleri gerçekleştirir:
*   $a = 1$ ise, Bob kendi kuantum bitinde $\mathsf{B}$ bit çevirme (veya $X$ kapısı) işlemi gerçekleştirir.
    * $b = 1$ ise, Bob kuantum bitinde $\mathsf{B}$ bir faz çevirme (veya $Z$ geçidi) gerçekleştirir.
    
Yani, $ab$'nin $00,$ $01,$ $10$ veya $11$ olmasına bağlı olarak, Bob kuantum bitinde $\mathbb{I},$ $Z,$ $X$ veya $ZX$ işlemlerinden birini gerçekleştirir.
Bu, ışınlanma protokolünün tam açıklamasıdır.
Aşağıda yer alan analiz, protokolün çalıştırılması durumunda, kuantum bitinin $\mathsf{B}$ protokolün yürütülmesinden önceki $\mathsf{Q}$ durumuna geçeceğini ortaya koymaktadır. Bu durum, diğer sistemlerle olan tüm korelasyonları da içermektedir. Yani protokol, $\mathsf{Q}$ durumunun $\mathsf{B}$'ye “ışınlandığı” mükemmel bir kuantum bit iletişim kanalını etkin bir şekilde uygulamıştır.
Analize geçmeden önce, bu protokolün $\mathsf{Q}$'nun durumunu klonlamada başarılı olmadığını unutmayın. Klonlama teoremine göre bunun imkansız olduğunu zaten biliyoruz.
Aksine, protokol tamamlandığında, kuantum bitinin $\mathsf{Q}$ durumu, üzerinde yapılan ölçümün sonucu olarak orijinal değerinden $\vert b\rangle$ değerine değişecektir.
Ayrıca, e-bit'in bu süreçte etkili bir şekilde “yakılmış” olduğunu da unutmayın: $\mathsf{A}$'nın durumu $\vert a\rangle$ olarak değişmiştir ve artık $\mathsf{B}$ (veya başka herhangi bir sistem) ile dolanık değildir.
Bu, ışınlanmanın bedelidir.
## Analiz
Işınlanma protokolünü analiz etmek için, yukarıda açıklanan devrenin davranışını, $\mathsf{Q}$'nun başlangıçta $\alpha\vert 0\rangle + \beta\vert 1\rangle$ durumunda olduğu durumdan başlayarak, adım adım inceleyeceğiz.
Bu, $\mathsf{Q}$'nun diğer sistemlerle dolanık olma olasılığını içermediğinden en genel durum değildir, ancak bu daha basit durumdan başlamak analizi daha net hale getirecektir.
Daha genel durum, daha basit durumun analizinden sonra aşağıda ele alınacaktır.
Özellikle, bu şekilde gösterilen zamanlarda kuantum bitlerinin $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumlarını ele alacağız:



Kubit $\mathsf{Q}$'nun protokolü $\alpha\vert 0\rangle + \beta\vert 1\rangle$ durumunda başladığı varsayıldığında, protokolün başlangıcında üç kubitin $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumu şu şekildedir
$$
\vert \pi_0 \rangle
= \vert \phi^+\rangle \otimes
\bigl(\alpha\vert 0\rangle + \beta\vert 1\rangle \bigr)
= \frac{\alpha \vert 000 \rangle + \alpha \vert 110\rangle + \beta \vert 001\rangle + \beta \vert 111\rangle}{\sqrt{2}}.
$$
Gerçekleştirilen ilk kapı, $\vert\pi_0\rangle$ durumunu
$$
\vert \pi_1 \rangle  = \frac{\alpha \vert 000 \rangle + \alpha \vert 110\rangle + \beta \vert 011\rangle + \beta \vert 101\rangle}{\sqrt{2}}.
$$
Ardından Hadamard kapısı uygulanır ve bu kapı, $\vert\pi_1\rangle$ durumunu şu şekilde dönüştürür
$$
\begin{aligned}
\vert\pi_2\rangle
& =
\frac{\alpha \vert 00\rangle \vert + \rangle + \alpha \vert 11\rangle\vert +\rangle + \beta \vert 01\rangle\vert -\rangle + \beta \vert 10\rangle\vert -\rangle}{\sqrt{2}}\\[2mm]
& = \frac{\alpha \vert 000 \rangle
+ \alpha \vert 001 \rangle
+ \alpha \vert 110 \rangle
+ \alpha \vert 111 \rangle
+ \beta \vert 010 \rangle
- \beta \vert 011 \rangle
+ \beta \vert 100 \rangle
- \beta \vert 101 \rangle}{2}.
\end{aligned}
$$
Tensör çarpımının çok doğrusal özelliğini kullanarak, bu durumu alternatif olarak şu şekilde yazabiliriz:
$$
\begin{aligned}
\vert\pi_2\rangle = \quad
& \frac{1}{2} \bigl(\alpha\vert 0 \rangle + \beta \vert 1\rangle \bigr)\vert 00\rangle \\[2mm]
+ & \frac{1}{2} \bigl(\alpha\vert 0 \rangle - \beta \vert 1\rangle \bigr)\vert 01\rangle \\[2mm]
+ & \frac{1}{2} \bigl(\alpha\vert 1 \rangle + \beta \vert 0\rangle \bigr)\vert 10\rangle \\[2mm]
+ & \frac{1}{2} \bigl(\alpha\vert 1 \rangle - \beta \vert 0\rangle \bigr)\vert 11\rangle.
\end{aligned}
$$
İlk bakışta, sanki sihirli bir şey olmuş gibi görünebilir, çünkü en soldaki kuantum biti $\mathsf{B}$ artık $\alpha$ ve $\beta$ sayılarına bağlı gibi görünüyor, ancak Alice'den Bob'a henüz herhangi bir iletişim olmamıştır.
Bu bir yanılsamadır.
Skalarlar tensör çarpımlarında serbestçe dolaşır, bu nedenle $\alpha$ ve $\beta$, diğer kuantum bitlerinden daha fazla veya daha az sol taraftaki kuantum bitine bağlı değildir ve tek yaptığımız, ölçümlerin analizini kolaylaştıracak şekilde durumu cebirsel olarak ifade etmektir.
Şimdi Alice'in standart baz ölçümlerinin dört olası sonucunu ve Bob'un bunun sonucunda gerçekleştirdiği eylemleri ele alalım.

### Olası sonuçlar

*   Alice'in ölçümünün sonucu, olasılıkla $ab = 00$'dır.

    $$
    \Biggl\| \frac{1}{2}\bigl(\alpha \vert 0\rangle + \beta\vert 1\rangle\bigr) \Biggr\|^2
    = \frac{\vert\alpha\vert^2 + \vert\beta\vert^2}{4} = \frac{1}{4},
    $$

    bu durumda $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumu şu şekilde olur

    $$
    \bigl( \alpha \vert 0 \rangle + \beta \vert 1 \rangle \bigr) \vert 00 \rangle.
    $$

    Bob bu durumda hiçbir şey yapmaz ve bu üç kubitin son durumu budur.

*   Alice'in ölçümünün sonucu, olasılıkla $ab = 01$'dir.

    $$
    \Biggl\| \frac{1}{2}\bigl(\alpha \vert 0\rangle - \beta\vert 1\rangle\bigr) \Biggr\|^2
    = \frac{\vert\alpha\vert^2 + \vert{-\beta}\vert^2}{4} = \frac{1}{4},
    $$
    in which case the state of $(\mathsf{B},\mathsf{A},\mathsf{Q})$ becomes
    $$
    \bigl( \alpha \vert 0 \rangle - \beta \vert 1 \rangle \bigr) \vert 01 \rangle.
    $$
    In this case Bob applies a $Z$ gate to $\mathsf{B},$ leaving $(\mathsf{B},\mathsf{A},\mathsf{Q})$ in the state
    $$
    \bigl( \alpha \vert 0 \rangle + \beta \vert 1 \rangle \bigr) \vert 01 \rangle.
    $$

*   Alice'in ölçümünün sonucu, olasılıkla $ab = 10$'dur.

    $$
    \Biggl\| \frac{1}{2}\bigl(\alpha \vert 1\rangle + \beta\vert 0\rangle\bigr) \Biggr\|^2
    = \frac{\vert\alpha\vert^2 + \vert\beta\vert^2}{4} = \frac{1}{4},
    $$
    
    Bu durumda $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumu
    $$
    \bigl( \alpha \vert 1 \rangle + \beta \vert 0 \rangle \bigr) \vert 10 \rangle olur.
    
    $$
    Bu durumda Bob, kuantum bitine $\mathsf{B}$ bir $X$ geçidi uygular ve $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumunu
    $$
    \bigl( \alpha \vert 0 \rangle + \beta \vert 1 \rangle \bigr) \vert 10 \rangle.
    $$

*   Alice'in ölçümünün sonucu, olasılıkla $ab = 11$'dir.

    $$
    \Biggl\| \frac{1}{2}\bigl(\alpha \vert 1\rangle - \beta\vert 0\rangle\bigr) \Biggr\|^2
    = \frac{\vert\alpha\vert^2 + \vert{-\beta}\vert^2}{4} = \frac{1}{4},
    $$
    
    Bu durumda $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumunun
    $$
    \bigl( \alpha \vert 1 \rangle - \beta \vert 0 \rangle \bigr) \vert 11 \rangle.
    
    $$
    Bu durumda Bob, kuantum bit $\mathsf{B}$ üzerinde $ZX$ işlemini gerçekleştirir ve $(\mathsf{B},\mathsf{A},\mathsf{Q})$ durumunu
    $$
    \bigl( \alpha \vert 0 \rangle + \beta \vert 1 \rangle \bigr) \vert 11 \rangle.

Şimdi, dört durumda da Bob'un kuantum bitinin $\mathsf{B}$ protokolün sonunda
$\alpha\vert 0\rangle + \beta\vert 1\rangle$ durumunda kaldığını görüyoruz, bu da kuantum bitinin $\mathsf{Q}$ başlangıç durumudur.
İşte göstermek istediğimiz şey buydu: ışınlanma protokolü doğru çalıştı.
Ayrıca, kuantum bitleri $\mathsf{A}$ ve $\mathsf{Q}$'nun, Alice'in elde ettiği ölçüm sonuçlarına bağlı olarak, her biri $1/4$ olasılıkla dört durumdan biri olan $\vert 00\rangle,$ $\vert 01\rangle,$ $\vert 10\rangle,$ veya $\vert 11\rangle$ durumunda kaldığını görüyoruz.
Böylece, yukarıda da belirtildiği gibi, protokolün sonunda Alice artık $\alpha \vert 0\rangle + \beta \vert 1\rangle$ durumuna sahip değildir, bu da klonlama teoremiyle tutarlıdır.
Alice'in ölçümlerinin $\alpha \vert 0\rangle + \beta \vert 1\rangle$ durumu hakkında kesinlikle hiçbir bilgi vermediğine dikkat edin.
Yani, dört olası ölçüm sonucunun her birinin olasılığı, $\alpha$ ve $\beta$'dan bağımsız olarak $1/4$'tür.
Bu, teleportasyonun doğru şekilde çalışması için de gereklidir.
Bilinmeyen bir kuantum durumundan bilgi çıkarmak genel olarak onu bozar, ancak burada Bob durumu bozulmadan elde eder.
Şimdi, kuantum bitinin $\mathsf{Q}$ başlangıçta başka bir sistemle dolanık olduğu daha genel bir durumu ele alalım. Bu sistemi $\mathsf{R}$ olarak adlandıralım.
Yukarıdakine benzer bir analiz, teleportasyon protokolünün bu daha genel durumda da doğru çalıştığını ortaya koyar:
protokolün sonunda, Bob'un elindeki kuantum bit $\mathsf{B}$ protokolün başlangıcında $\mathsf{Q}$'nun olduğu gibi $\mathsf{R}$ ile dolanık hale gelir, sanki Alice $\mathsf{Q}$'yu Bob'a vermiş gibi.
Bunu kanıtlamak için, $(\mathsf{Q},\mathsf{R})$ çiftinin durumunun başlangıçta aşağıdaki formdaki bir kuantum durum vektörü ile verildiğini varsayalım
$$
\alpha \vert 0 \rangle_{\mathsf{Q}} \vert \gamma_0\rangle_{\mathsf{R}}
+ \beta \vert 1 \rangle_{\mathsf{Q}} \vert \gamma_1\rangle_{\mathsf{R}},
$$
burada $\vert\gamma_0\rangle$ ve $\vert\gamma_1\rangle$ sistem $\mathsf{R}$ için kuantum durum vektörleri ve $\alpha$ ve $\beta$ $\vert \alpha \vert^2 + \vert\beta\vert^2 = 1$ koşulunu sağlayan karmaşık sayılardır.
Çiftin $(\mathsf{Q},\mathsf{R})$ herhangi bir kuantum durum vektörü bu şekilde ifade edilebilir.
Aşağıdaki şekil, önceki devreye $\mathsf{R}$ sisteminin eklenmesiyle (diyagramın üstünde hiçbir şeyin olmadığı kuantum bitlerinin bir koleksiyonu ile temsil edilir) aynı devreyi göstermektedir.



Işınlanma protokolü çalıştırıldığında neler olduğunu analiz etmek için, önceki derste anlatıldığı gibi sistemleri permütasyonuna tabi tutmak yararlı olacaktır.
Özellikle, sistemlerin durumunu $(\mathsf{B},\mathsf{R},\mathsf{A},\mathsf{Q})$ sırasıyla ele alacağız, $(\mathsf{B},\mathsf{A},\mathsf{Q},\mathsf{R})$ sırasıyla değil.
Çeşitli sistemlerin isimleri, açıklık sağlamak için aşağıdaki ifadelerde alt simge olarak verilmiştir.
Protokolün başlangıcında, bu sistemlerin durumu aşağıdaki gibidir:
$$
\begin{aligned}
\vert \pi_0\rangle
& = \vert \phi^+\rangle_{\mathsf{BA}} \otimes \bigl(
  \alpha \vert 0\rangle_{\mathsf{Q}} \vert\gamma_0\rangle_{\mathsf{R}}
+ \beta \vert 1\rangle_{\mathsf{Q}}\vert\gamma_1\rangle_{\mathsf{R}}\bigr)\\[1mm]
& = \frac{
\alpha \vert 0\rangle_{\mathsf{B}} \vert \gamma_0 \rangle_{\mathsf{R}} \vert 00 \rangle_{\mathsf{AQ}}
+ \alpha \vert 1\rangle_{\mathsf{B}} \vert \gamma_0 \rangle_{\mathsf{R}} \vert 10 \rangle_{\mathsf{AQ}}
+ \beta \vert 0\rangle_{\mathsf{B}} \vert \gamma_1 \rangle_{\mathsf{R}} \vert 01 \rangle_{\mathsf{AQ}}
+ \beta \vert 1\rangle_{\mathsf{B}} \vert \gamma_1 \rangle_{\mathsf{R}} \vert 11 \rangle_{\mathsf{AQ}}}{\sqrt{2}}.
\end{aligned}
$$
İlk olarak, bu durumu aşağıdaki hale dönüştüren kontrollü NOT kapısı uygulanır
$$
\vert\pi_1\rangle =
\frac{
  \alpha \vert 0\rangle_{\mathsf{B}} \vert\gamma_0 \rangle_{\mathsf{R}} \vert 00\rangle_{\mathsf{AQ}}
+ \alpha \vert 1\rangle_{\mathsf{B}} \vert\gamma_0 \rangle_{\mathsf{R}} \vert 10\rangle_{\mathsf{AQ}}
+ \beta \vert 0\rangle_{\mathsf{B}} \vert\gamma_1 \rangle_{\mathsf{R}} \vert 11\rangle_{\mathsf{AQ}}
+ \beta \vert 1\rangle_{\mathsf{B}} \vert\gamma_1 \rangle_{\mathsf{R}} \vert 01\rangle_{\mathsf{AQ}}}{\sqrt{2}}.
$$
Ardından Hadamard kapısı uygulanır.
Ortaya çıkan durumu genişletip basitleştirdikten sonra, yukarıdaki daha basit durumun analizine benzer şekilde, ortaya çıkan durumun şu ifadesini elde ederiz:
$$
\begin{aligned}
\vert \pi_2 \rangle = \quad
  & \frac{1}{2} \bigl(
   \alpha \vert 0\rangle_{\mathsf{B}} \vert\gamma_0\rangle_{\mathsf{R}}
  
+ \beta \vert 1\rangle_{\mathsf{B}} \vert\gamma_1\rangle_{\mathsf{R}}
  \bigr) \vert 00\rangle_{\mathsf{AQ}}\\[2mm]
  
+ & \frac{1}{2} \bigl(
   \alpha \vert 0\rangle_{\mathsf{B}} \vert\gamma_0\rangle_{\mathsf{R}}
- \beta \vert 1\rangle_{\mathsf{B}} \vert\gamma_1\rangle_{\mathsf{R}}
  
\bigr) \vert 01\rangle_{\mathsf{AQ}}\\[2mm]
  + & \frac{1}{2} \bigl(
   \alpha \vert 1\rangle_{\mathsf{B}} \vert\gamma_0\rangle_{\mathsf{R}}
  
+ \beta \vert 0\rangle_{\mathsf{B}} \vert\gamma_1\rangle_{\mathsf{R}}
  \bigr) \vert 10\rangle_{\mathsf{AQ}}\\[2mm]
  
+ & \frac{1}{2} \bigl(
   \alpha \vert 1\rangle_{\mathsf{B}} \vert\gamma_0\rangle_{\mathsf{R}}
- \beta \vert 0\rangle_{\mathsf{B}} \vert\gamma_1\rangle_{\mathsf{R}}
  
\bigr) \vert 11\rangle_{\mathsf{AQ}}.
\end{aligned}
$$
Daha önce olduğu gibi, Alice'in ölçümlerinin dört farklı olası sonucunu ve Bob'un gerçekleştirdiği ilgili eylemleri dikkate alarak, protokolün sonunda $(\mathsf{B},\mathsf{R})$ durumunun her zaman
$$
\alpha \vert 0 \rangle \vert \gamma_0\rangle + \beta \vert 1 \rangle \vert \gamma_1\rangle.
$$
Gayri resmi olarak konuşursak, analiz yukarıdaki daha basit duruma kıyasla önemli bir şekilde değişmez;
$\vert\gamma_0\rangle$ ve $\vert\gamma_1\rangle$ esasen sadece “yolculuğa eşlik eder”.
Böylece, teleportasyon mükemmel bir kuantum iletişim kanalı oluşturmayı başarır, kuantum bitinin $\mathsf{Q}$ içeriğini $\mathsf{B}$'ye etkili bir şekilde aktarır ve diğer sistemlerle olan tüm korelasyonları korur.
Yukarıdaki daha basit durumun analizine bakıldığında, bu aslında hiç de şaşırtıcı değildir.
Bu analizde ortaya çıktığı gibi, rastgele bir kuantum durumunda bir kuantum bitinde kimlik işlemi gibi davranan fiziksel bir süreç vardır ve bunun gerçekleşmesinin tek bir yolu vardır: protokol tarafından uygulanan işlem kimlik işlemi *olmalıdır*.
Yani, tek bir kuantum bit için izole olarak teleportasyonun doğru çalıştığını bildiğimizde, protokolün etkili bir şekilde mükemmel, gürültüsüz bir kuantum kanalı uyguladığı sonucuna varabiliriz ve bu nedenle, giriş kuantum biti başka bir sistemle dolanık olsa bile protokolün doğru çalışması gerekir.
## Daha fazla tartışma
Teleportasyon hakkında birkaç kısa sonuç yorumunu aşağıda bulabilirsiniz.
İlk olarak, teleportasyon kuantum bilgisinin bir *uygulaması* değil, kuantum iletişimi gerçekleştirmek için bir *protokol*'dür.
Bu nedenle, kuantum iletişiminin yararlı olduğu ölçüde yararlıdır.
Gerçekten de, teleportasyonun bir gün, belki de *dolaşıklık damıtma* olarak bilinen bir süreç aracılığıyla, kuantum bilgisini iletmek için standart bir yol haline gelebileceğini düşünmek mantıklıdır.
Bu süreç, daha fazla sayıda gürültülü (veya kusurlu) e-bit'i daha az sayıda yüksek kaliteli e-bit'e dönüştürür ve bu e-bit'ler gürültüsüz veya neredeyse gürültüsüz teleportasyon için kullanılabilir.
Buradaki fikir, dolanıklık damıtma sürecinin doğrudan kuantum iletişimi kadar hassas olmadığıdır.
Örneğin, kayıpları kabul edebiliriz ve süreç işe yaramazsa tekrar deneyebiliriz.
Buna karşılık, iletmeyi umduğumuz gerçek kuantum bitleri çok daha değerli olabilir.
Son olarak, ışınlanmanın ardındaki fikir ve çalışma şeklinin kuantum bilgi ve hesaplamada oldukça temel olduğu anlaşılmalıdır.
Bu, kuantum bilgi teorisinin gerçekten bir köşe taşıdır ve bunun çeşitli varyasyonları ortaya çıkmaktadır.
Örneğin, kuantum kapıları, kuantum kapı ışınlaması olarak bilinen ve ışınlamayı kuantum bitlerine iletişim kurmak yerine *işlemler* uygulamak için kullanan, yakından ilişkili bir süreçle uygulanabilir.

```python
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
from qiskit.circuit.library import Initialize
from qiskit.result import marginal_counts
from qiskit.quantum_info import random_statevector
```

```python
## SETUP
# Protocol uses 3 qubits and 2 classical bits in 2 different registers

qr = QuantumRegister(3, name="q")    # Protocol uses 3 qubits
crz = ClassicalRegister(1, name="crz") # and 2 classical bits
crx = ClassicalRegister(1, name="crx") # in 2 different registers
teleportation_circuit = QuantumCircuit(qr, crz, crx)
```

#### Adım 1
Üçüncü bir taraf olan Telamon, birbiriyle dolanık bir kuantum bit çifti oluşturur ve birini Bob'a, diğerini Alice'e verir.
Telamon'un oluşturduğu çift, Bell çifti adı verilen özel bir çifttir. Kuantum devre dilinde, iki kuantum bit arasında Bell çifti oluşturmanın yolu, önce Hadamard kapısı kullanarak bunlardan birini X-temeline ($|+\rangle$ ve $|-\rangle$) aktarmak, ardından X-temelinde kontrol edilen diğer kuantum bitine CNOT kapısı uygulamaktır.

```python
def create_bell_pair(qc, a, b):
    """Creates a bell pair in qc using qubits a & b"""
    qc.h(a) # Put qubit a into state |+>
    qc.cx(a,b) # CNOT with a as control and b as target
```

```python
## SETUP
# Protocol uses 3 qubits and 2 classical bits in 2 different registers
qr = QuantumRegister(3, name="q")
crz, crx = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx")
teleportation_circuit = QuantumCircuit(qr, crz, crx)

## STEP 1
# In our case, Telamon entangles qubits q1 and q2
# Let's apply this to our circuit:
create_bell_pair(teleportation_circuit, 1, 2)
# And view the circuit so far:
teleportation_circuit.draw(output='mpl')
```

Alice ve Bob ayrıldıktan sonra Alice'in $q_1$ ve Bob'un $q_2$'ye sahip olduğunu varsayalım.
#### Adım 2
Alice, $\vert\psi\rangle$ (Bob'a göndermeye çalıştığı kuantum biti) tarafından kontrol edilen $q_1$'e bir CNOT kapısı uygular. Ardından Alice, $|\psi\rangle$'ye bir Hadamard kapısı uygular. Kuantum devremizde, Alice'in göndermeye çalıştığı kuantum biti ($|\psi\rangle$) $q_0$'dır:

```python
def alice_gates(qc, psi, a):
    qc.cx(psi, a)
    qc.h(psi)
```

```python
## SETUP
# Protocol uses 3 qubits and 2 classical bits in 2 different registers
qr = QuantumRegister(3, name="q")
crz, crx = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx")
teleportation_circuit = QuantumCircuit(qr, crz, crx)

## STEP 1
create_bell_pair(teleportation_circuit, 1, 2)

## STEP 2
teleportation_circuit.barrier() # Use barrier to separate steps
alice_gates(teleportation_circuit, 0, 1)
teleportation_circuit.draw(output='mpl')
```

#### Adım 3
Ardından Alice, sahip olduğu her iki kuantum bitine, $q_1$ ve $\vert\psi\rangle$, bir ölçüm uygular ve bu sonucu iki klasik bite kaydeder. Daha sonra bu iki biti Bob'a gönderir.

```python
def measure_and_send(qc, a, b):
    """Measures qubits a & b and 'sends' the results to Bob"""
    qc.barrier()
    qc.measure(a,0)
    qc.measure(b,1)
```

```python
## SETUP
# Protocol uses 3 qubits and 2 classical bits in 2 different registers
qr = QuantumRegister(3, name="q")
crz, crx = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx")
teleportation_circuit = QuantumCircuit(qr, crz, crx)

## STEP 1
create_bell_pair(teleportation_circuit, 1, 2)

## STEP 2
teleportation_circuit.barrier() # Use barrier to separate steps
alice_gates(teleportation_circuit, 0, 1)

## STEP 3
measure_and_send(teleportation_circuit, 0 ,1)
teleportation_circuit.draw(output='mpl')
```

#### Adım 4
Zaten $q_2$ kuantum bitine sahip olan Bob, klasik bitlerin durumuna bağlı olarak aşağıdaki geçitleri uygular:
00 $\rightarrow$ Hiçbir şey yapma
01 $\rightarrow$ $X$ geçidi uygula
10 $\rightarrow$ $Z$ geçidi uygula
11 $\rightarrow$ $ZX$ geçidi uygula
(*Bu bilgi aktarımının tamamen klasik olduğunu unutmayın*.)

```python
def bob_gates(qc, qubit, crz_bit, crx_bit):
    # crx_bit ve crz_bit tek tek Clbit nesneleri (ya da 1-bitlik ClassicalRegister)
    with qc.if_test((crx_bit, 1)):   # crx == 1 ise
        qc.x(qubit)
    with qc.if_test((crz_bit, 1)):   # crz == 1 ise
        qc.z(qubit)
```

```python
## SETUP
# Protocol uses 3 qubits and 2 classical bits in 2 different registers
qr = QuantumRegister(3, name="q")
crz, crx = ClassicalRegister(1, name="crz"), ClassicalRegister(1, name="crx")
teleportation_circuit = QuantumCircuit(qr, crz, crx)

## STEP 1
create_bell_pair(teleportation_circuit, 1, 2)

## STEP 2
teleportation_circuit.barrier() # Use barrier to separate steps
alice_gates(teleportation_circuit, 0, 1)

## STEP 3
measure_and_send(teleportation_circuit, 0, 1)

## STEP 4
teleportation_circuit.barrier() # Use barrier to separate steps
bob_gates(teleportation_circuit, 2, crz, crx)
teleportation_circuit.draw(output='mpl')
```

## 3. Işınlanma Protokolünün Simülasyonu <a id=‘simulating’></a>

### 3.1 Protokolü Kuantum Bilgisayarda Nasıl Test Edeceğiz? <a id=‘testing’></a>

Bu not defterinde, Alice'in kubitini rastgele bir durum olan $\vert\psi\rangle$ (`psi`) ile başlatacağız. Bu durum, $|q_0\rangle$ üzerinde bir `Initialize` geçidi kullanılarak oluşturulacaktır. Bu bölümde, `psi`'yi seçmek için `random_statevector` işlevini kullanıyoruz, ancak `psi`'yi istediğiniz herhangi bir kubit durumuna ayarlayabilirsiniz.

```python
# Create random 1-qubit state
psi = random_statevector(2)

# Display it nicely
display(array_to_latex(psi, prefix="|\\psi\\rangle ="))
# Show it on a Bloch sphere
plot_bloch_multivector(psi)
```

```python
init_gate = Initialize(psi)
init_gate.label = "init"
```

(`Initialize` teknik olarak bir geçit değildir, çünkü bir sıfırlama işlemi içerir ve bu nedenle geri döndürülemez. Biz buna ‘komut’ diyoruz). Kuantum ışınlama devresi çalışırsa, devrenin sonunda kuantum biti $|q_2\rangle$ bu durumda olacaktır. Bunu durum vektörü simülatörü kullanarak kontrol edeceğiz.

```python
## SETUP
qr = QuantumRegister(3, name="q")   # Protocol uses 3 qubits
crz = ClassicalRegister(1, name="crz") # and 2 classical registers
crx = ClassicalRegister(1, name="crx")
qc = QuantumCircuit(qr, crz, crx)

## STEP 0
# First, let's initialize Alice's q0
qc.append(init_gate, [0])
qc.barrier()

## STEP 1
# Now begins the teleportation protocol
create_bell_pair(qc, 1, 2)
qc.barrier()

## STEP 2
# Send q1 to Alice and q2 to Bob
alice_gates(qc, 0, 1)

## STEP 3
# Alice then sends her classical bits to Bob
measure_and_send(qc, 0, 1)

## STEP 4
# Bob decodes qubits
bob_gates(qc, 2, crz, crx)

# Display the circuit
qc.draw(output='mpl')
```

Aşağıda, aer simülatöründen elde edilen durum vektörünü kullanarak, $|q_2\rangle$ durumunun yukarıda oluşturduğumuz $|\psi\rangle$ durumu ile aynı olduğunu, $|q_0\rangle$ ve $|q_1\rangle$ durumlarının ise $|0\rangle$ veya $|1\rangle$ durumlarına çöktüğünü görebiliriz. $|\psi\rangle$ durumu, kuantum bit 0'dan kuantum bit 2'ye ışınlanmıştır.

```python
sim = AerSimulator()
qc.save_statevector()
out_vector = sim.run(qc).result().get_statevector()
plot_bloch_multivector(out_vector)
```

Emin olmak için bu hücreyi birkaç kez çalıştırabilirsiniz. Qubit 0 ve 1'in durumlarının değiştiğini, ancak qubit 2'nin her zaman $|\psi\rangle$ durumunda olduğunu fark edebilirsiniz.

### 3.3 Simüle Edilen Sayımları Kullanma <a id=‘simulating-fc’></a>
Kuantum ışınlama, iki taraf arasında kuantum bitleri göndermek için tasarlanmıştır. Bunu gösterecek donanımımız yoktur, ancak kapıların tek bir kuantum çip üzerinde doğru dönüşümleri gerçekleştirdiğini gösterebiliriz. Burada yine aer simülatörünü kullanarak protokolümüzü nasıl test edebileceğimizi simüle edeceğiz.
Gerçek bir kuantum bilgisayarda, durum vektörünü örnekleyemeyiz, bu nedenle teleportasyon devremizin çalıştığını kontrol etmek istiyorsak, biraz farklı bir yol izlememiz gerekir. `Initialize` komutu önce bir sıfırlama gerçekleştirir ve kuantum bitimizi $|0\rangle$ durumuna ayarlar. Ardından, $|0\rangle$ kuantum bitimizi $|\psi\rangle$ durumuna dönüştürmek için kapılar uygular:
$$ |0\rangle \xrightarrow{\text{Initialize gates}} |\psi\rangle $$
Tüm kuantum kapılar tersine çevrilebilir olduğundan, bu kapıların tersini şu şekilde bulabiliriz:

```python
inverse_init_gate = init_gate.gates_to_uncompute()
```

Bu işlem şu özelliğe sahiptir:
$$ |\psi\rangle \xrightarrow{\text{Ters Başlatma kapıları}} |0\rangle $$
Kubit $|q_0\rangle$'nin $|q_2\rangle$'ye ışınlandığını kanıtlamak için, $|q_2\rangle$ üzerinde bu ters başlatmayı yaparsak, kesin olarak $|0\rangle$ ölçümünü almayı bekleriz. Bunu aşağıdaki devrede yapıyoruz:

```python
## SETUP
qr = QuantumRegister(3, name="q")   # Protocol uses 3 qubits
crz = ClassicalRegister(1, name="crz") # and 2 classical registers
crx = ClassicalRegister(1, name="crx")
qc = QuantumCircuit(qr, crz, crx)

## STEP 0
# First, let's initialize Alice's q0
qc.append(init_gate, [0])
qc.barrier()

## STEP 1
# Now begins the teleportation protocol
create_bell_pair(qc, 1, 2)
qc.barrier()

## STEP 2
# Send q1 to Alice and q2 to Bob
alice_gates(qc, 0, 1)

## STEP 3
# Alice then sends her classical bits to Bob
measure_and_send(qc, 0, 1)

## STEP 4
# Bob decodes qubits
bob_gates(qc, 2, crz, crx)

## STEP 5
# reverse the initialization process
qc.append(inverse_init_gate, [2])

# Display the circuit
qc.draw(output='mpl', fold=-1)
```

Devre şemasında ‘disentangler’ olarak etiketlenmiş `inverse_init_gate`'in ortaya çıktığını görebiliriz. Son olarak, üçüncü kuantum bitini ölçer ve sonucu üçüncü klasik bite kaydederiz:

```python
# Need to add a new ClassicalRegister
# to see the result
cr_result = ClassicalRegister(1)
qc.add_register(cr_result)
qc.measure(2,2)
qc.draw(output='mpl', fold=-1)
```

```python
t_qc = transpile(qc, sim)
t_qc.save_statevector()
counts = sim.run(t_qc).result().get_counts()
qubit_counts = [marginal_counts(counts, [qubit]) for qubit in range(3)]
plot_histogram(qubit_counts)
```

$q_2$'yi (histogramdaki mor çubuk) $|0\rangle$ durumunda %100 olasılıkla ölçebileceğimizi görüyoruz. Bu beklenen sonuçtur ve teleportasyon protokolünün düzgün çalıştığını gösterir.
