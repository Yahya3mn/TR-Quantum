# Süper Yoğun Kodlama
Süper yoğun kodlama, bir anlamda teleportasyonun tamamlayıcı bir amacını gerçekleştiren bir protokoldür.
İki klasik bitlik iletişim kullanarak bir kuantum bitinin iletilmesine izin vermek yerine (bir e-bitlik dolanıklık pahasına), bir kuantum bit kullanarak iki klasik bitin iletilmesine izin verir (yine bir e-bitlik dolanıklık pahasına).
Daha ayrıntılı olarak, bir e-bit dolanıklığı paylaşan bir gönderici (Alice) ve bir alıcı (Bob) vardır.
Ders için geçerli olan kurallara göre, bu, Alice'in bir kuantum biti $\mathsf{A}$'ya, Bob'un bir kuantum biti $\mathsf{B}$'ye sahip olduğu ve $(\mathsf{A},\mathsf{B})$ çiftinin birlikte $\vert\phi^+\rangle$ durumunda olduğu anlamına gelir.
Alice, Bob'a iki klasik bit iletmek istemektedir; bunları $c$ ve $d$ ile göstereceğiz. Bunu, Bob'a bir kuantum bit göndererek gerçekleştirecektir.
Bu başarıyı, teleportasyonun başardığından daha az ilginç olarak görmek mantıklıdır.
Kubitleri göndermek, öngörülebilir gelecekte klasik bitleri göndermekten çok daha zor olacağından, bir kuantum iletişimi kubiti karşılığında iki klasik iletişim biti elde etmek, bir e-bit bedeli ödeyerek, pek de değecek bir şey gibi görünmemektedir.
Ancak bu, süper yoğun kodlamanın ilginç olmadığı anlamına gelmez, çünkü kesinlikle ilginçtir.
Dersin temasına uygun olarak, süper yoğun kodlamanın ilginç olmasının bir nedeni, dolanıklığın somut ve (bilgi teorisi bağlamında) oldukça çarpıcı bir kullanımını göstermesidir.
Kuantum bilgi teorisinde *Holevo teoremi* olarak bilinen ünlü bir teorem, paylaşılan dolanık durum kullanılmadan tek bir kuantum bit göndererek birden fazla klasik bilgi bitini iletmenin imkansız olduğunu ima eder.
(Holevo teoremi bundan daha geneldir.
Kesin ifadesi teknik ve açıklama gerektirir, ancak bu, teoremin sonuçlarından biridir.
Dolayısıyla, süper yoğun kodlama yoluyla, paylaşılan dolanıklık, kuantum bitleri göndermenin klasik bilgi taşıma kapasitesinin *iki katına çıkmasını* sağlar.
## Protokol
Aşağıdaki kuantum devre şeması, süper yoğun kodlama protokolünü açıklamaktadır:



Alice'in yaptığı şey şu şekilde açıklanabilir:
1.  $d=1$ ise, Alice kuantum bitinde $\mathsf{A}$ üzerinde bir $Z$ kapısı gerçekleştirir (ve $d=0$ ise gerçekleştirmez).
2.  $c=1$ ise, Alice kuantum bitinde $\mathsf{A}$ üzerinde bir $X$ kapısı gerçekleştirir (ve $c=0$ ise gerçekleştirmez).
Alice daha sonra kuantum bitini $\mathsf{A}$ Bob'a gönderir.
Bob, kuantum bitini $\mathsf{A}$ aldığında, önce $\mathsf{A}$'yı kontrol ve $\mathsf{B}$'yi hedef olarak kullanarak kontrollü NOT kapısı gerçekleştirir, ardından $\mathsf{A}$'ya Hadamard kapısı uygular.
Ardından, her iki durumda da standart temel ölçümler kullanarak $\mathsf{B}$'yi ölçerek $c$'yi ve $\mathsf{A}$'yı ölçerek $d$'yi elde eder.
## Analiz
Bu protokolün ardındaki fikir oldukça basittir:
Alice, Bob ile paylaşmak istediği Bell durumunu etkili bir şekilde seçer,
Bob'a kuantum bitini gönderir ve Bob, Alice'in hangi Bell durumunu seçtiğini belirlemek için ölçüm yapar.
Yani, başlangıçta $\vert\phi^+\rangle$'yi paylaşırlar ve $c$ ve $d$ bitlerine bağlı olarak Alice bu durumu olduğu gibi bırakır veya kuantum bitine $\mathbb{I},$ $X,$ $Z$ veya $XZ$ uygulayarak diğer Bell durumlarından birine geçirir.
$\mathsf{A}.$
$$
\begin{aligned}
(\mathbb{I} \otimes \mathbb{I}) \vert \phi^+ \rangle & = \vert \phi^+\rangle \\
(\mathbb{I} \otimes Z) \vert \phi^+ \rangle & = \vert \phi^-\rangle \\
(\mathbb{I} \otimes X) \vert \phi^+ \rangle & = \vert \psi^+\rangle \\
(\mathbb{I} \otimes XZ) \vert \phi^+ \rangle & = \vert \psi^-\rangle
\end{aligned}
$$
Bob'un eylemleri dört Bell durumuna aşağıdaki etkileri yapar:
$$
\begin{aligned}
\vert \phi^+\rangle & \mapsto \vert 00\rangle\\
\vert \phi^-\rangle & \mapsto \vert 01\rangle\\
\vert \psi^+\rangle & \mapsto \vert 10\rangle\\
\vert \psi^-\rangle & \mapsto -\vert 11\rangle\\
\end{aligned}
$$
Bu, Bob'un bu durumlar üzerinde yaptığı işlemlerin sonuçlarını tek tek hesaplayarak doğrudan kontrol edilebilir.
Böylece, Bob ölçümlerini yaptığında, Alice'in hangi Bell durumunu seçtiğini belirleyebilir.
Protokolün doğru çalıştığını doğrulamak için her durumu kontrol etmek yeterlidir:
*   Eğer $cd = 00,$ ise Bob $\mathsf{A}$'yı aldığında $(\mathsf{B},\mathsf{A})$'nın durumu $\vert \phi^+\rangle$ olur. O bu durumu $\vert 00\rangle$'ye dönüştürür ve $cd = 00$ elde eder.
* $cd = 01$ ise, Bob $\mathsf{A}$'yı aldığında $(\mathsf{B},\mathsf{A})$'nın durumu $\vert \phi^-\rangle$ olur. Bu durumu $\vert 01\rangle$'ye dönüştürür ve $cd = 01$ elde eder.
* $cd = 10$ ise, Bob $\mathsf{A}$'yı aldığında $(\mathsf{B},\mathsf{A})$'nın durumu $\vert \psi^+\rangle$ olur. Bu durumu $\vert 10\rangle$'ye dönüştürür ve $cd = 10$ elde eder.
* $cd = 11$ ise, Bob $\mathsf{A}$'yı aldığında $(\mathsf{B},\mathsf{A})$'nın durumu $\vert \psi^-\rangle$ olur. Bu durumu $-\vert 11\rangle$'ye dönüştürür ve $cd = 11$ elde eder. (Negatif bir faz faktörü burada etkili değildir.)


## 3. Süper Yoğun Kodlama Protokolünün Simülasyonu <a id=‘simulating’></a>

```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

Bir dolanık çift oluşturmak için H-kapısı ve ardından CNOT yapmamız gerektiğini gördük. `QuantumCircuit` alan ve `a` ve `b` indisli kuantum bitlerini dolanık hale getiren bir fonksiyon oluşturalım:

```python
def create_bel_pair():
    qc = QuantumCircuit(2)
    qc.h(1)  # Apply Hadamard gate to the first qubit
    qc.cx(1, 0)  # Apply CNOT gate with qubit 1 as control and qubit 0 as target
    return qc
```

Şimdi mesajımızı kodlamamız gerekiyor. Gönderilebilecek dört mesaj olduğunu gördük: `00`, `10`, `01` veya `11`. Bu mesajı alan ve bizim için uygun kapıları uygulayan bir fonksiyon oluşturalım:

```python
def encode_message(qc, qubit, msg):
    if len(msg) != 2 or not set(msg).issubset({"0","1"}):
        raise ValueError(f"message '{msg}' is invalid")
    if msg[1] == "1":
        qc.x(qubit)
    if msg[0] == "1":
        qc.z(qubit)
    return qc
```

Son olarak, mesajımızı çözmemiz gerekiyor. Bunu bir CNOT ve ardından bir H-kapısı kullanarak yapabileceğimizi gördük. Bunu bizim için yapan bir fonksiyon da oluşturalım:

```python
def decode_message(qc):
    qc.cx(1,0)
    qc.h(1)
    return qc
```

Son olarak, bunları bir araya getirerek protokolümüzü tamamlayabiliriz.

```python
qc = create_bel_pair()

qc.barrier()

message = "10"
qc = encode_message(qc, 1, message)
qc.barrier()

qc = decode_message(qc)

qc.measure_all()

qc.draw(output='mpl')
```

```python
aer_sim = AerSimulator()
result = aer_sim.run(qc).result()
counts = result.get_counts(qc)
print(counts)
plot_histogram(counts)
```

Simülatörümüz mükemmel bir kuantum bilgisayarı simüle eder. Hata olmadan, doğru mesajı ölçme şansımızın %100 olduğunu görebiliriz.
