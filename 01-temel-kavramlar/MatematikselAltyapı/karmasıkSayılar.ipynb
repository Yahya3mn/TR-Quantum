## Karmaşık Sayılar  

Kuantum mekaniğinin matematiksel yapısında **karmaşık sayılar** merkezi bir rol oynar.  

### 1. Tanım  

Bir karmaşık sayı, reel ve sanal kısımlardan oluşur:  

$$
z = a + i b, \quad a, b \in \mathbb{R}, \quad i^2 = -1
$$  

Burada $a$ reel kısım, $b$ ise sanal kısımdır.  

### 2. Karmaşık Düzlem  

Karmaşık sayılar, geometrik olarak **karmaşık düzlemde** gösterilir.  
- Reel kısım yatay eksene,  
- Sanal kısım dikey eksene karşılık gelir.  

Böylece her karmaşık sayı, iki boyutlu düzlemde bir nokta veya vektör gibi düşünülebilir.  

Bir karmaşık sayının büyüklüğü (modülü):  

$$
|z| = \sqrt{a^2 + b^2}
$$  

dir.  

### 3. Kuantum Mekaniğinde Kullanımı  

- Kuantum durumlarını tanımlayan dalga fonksiyonları genellikle **karmaşık değerli**dir:  

$$
\psi(x,t) \in \mathbb{C}
$$  

- Bunun nedeni, olasılık genliklerinin faz bilgisi taşımasıdır. Ölçüm olasılığı mutlak değer karesi ile hesaplanır:  

$$
P(x,t) = |\psi(x,t)|^2
$$  

- Faz (karmaşık kısmın etkisi), girişim ve süperpozisyon gibi kuantuma özgü fenomenlerde kritik rol oynar.

## İç Çarpım ve Hilbert Uzayları  

Kuantum mekaniğinde durum vektörlerinin yaşadığı ortam, sıradan bir vektör uzayı değil, özel bir yapı olan **Hilbert uzayı**dır.  

### 1. İç Çarpım  

Bir Hilbert uzayında temel yapı **iç çarpım**dır. İki vektör $u, v \in \mathbb{C}^n$ için iç çarpım şöyle tanımlanır:  

$$
\langle u, v \rangle = \sum_{i=1}^n u_i^* v_i
$$  

Burada $u_i^*$ karmaşık eşleniktir.  

İç çarpım sayesinde:  
- Norm (uzunluk) tanımlanır:  

$$
\|v\| = \sqrt{\langle v, v \rangle}
$$  

- Ortogonallik (diklik) tanımlanır:  

$$
\langle u, v \rangle = 0 \quad \Rightarrow \quad u \perp v
$$  

### 2. Hilbert Uzayı  

Bir **Hilbert uzayı**, iç çarpımlı ve tam bir vektör uzayıdır.  
- “Tam” demek, Cauchy dizilerinin limitlerinin uzayın içinde olmasıdır.  
- Bu özellik, analizde süreklilik ve yakınsama açısından büyük önem taşır.  

### 3. Kuantum Durumları  

- Kuantum sistemlerinin durumları, Hilbert uzayının birer vektörüdür:  

$$
|\psi\rangle \in \mathcal{H}
$$  

- Ölçüm sonuçlarının olasılıkları iç çarpım üzerinden hesaplanır:  

$$
P(\phi|\psi) = |\langle \phi | \psi \rangle|^2
$$  

- Hermit operatörler (gözlemlenebilirler), Hilbert uzayı üzerinde tanımlanır ve özvektör–özdeğer yapısı ölçümün matematiksel temelini oluşturur.  

---

### Genel Özet  

- Karmaşık sayılar: Kuantum genliklerini tanımlar, faz bilgisi içerir.  
- İç çarpım: Uzunluk ve açı kavramlarını sağlar, olasılıklarla doğrudan ilişkilidir.  
- Hilbert uzayı: Kuantum durumlarının tanımlandığı, iç çarpımlı ve tam bir vektör uzayıdır.
