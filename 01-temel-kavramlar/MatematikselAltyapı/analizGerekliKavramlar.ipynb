## Analizden Gerekli Kavramlar  

### Öğrenme Hedefleri  
Bu bölümü tamamladıktan sonra:  
- Limit ve sürekliliğin resmi tanımını verir, dalga fonksiyonlarına uygulayabilirsiniz.  
- Türev ve integral kavramlarını kuantumdaki karşılıklarıyla ilişkilendirip hesap yapabilirsiniz.  
- Fourier dönüşümünü tanımlayıp konum–momentum uzayları arasındaki ilişkiyi kurabilir, temel özelliklerini kullanabilirsiniz.  

---

### Limit ve Süreklilik  

**Tanım (Limit).**  
Bir fonksiyon için  

$$
\lim_{x\to a} f(x) = L
$$  

demek, $x$ değeri $a$’ya yaklaştığında $f(x)$ değerlerinin $L$’e yaklaşmasıdır.  

**Tanım (Süreklilik).**  
Fonksiyon $f$, $x=a$’da süreklidir iff  

$$
\lim_{x\to a} f(x) = f(a).
$$  

**Kuantumdaki Rolü.**  
- Schrödinger denklemi:  

$$
-\frac{\hbar^2}{2m}\,\psi''(x)+V(x)\,\psi(x)=E\,\psi(x)
$$  

diferansiyel olduğu için $\psi(x)$ çoğu durumda sürekli olmalıdır.  
- Potansiyel sıçramalarında $\psi$ süreklidir, fakat $\psi'$ süreksiz olabilir (ör. delta potansiyelinde).  

**Örnek**  

$$
\lim_{x\to 0}\frac{\sin x}{x} = 1
$$  

**Örnek**  

$$
f(x)=
\begin{cases}
x^2-1, & x<1 \\
2x-2, & x\ge 1
\end{cases}
$$  

$x=1$’de sağ ve sol limitler eşit ve $f(1)=0$, dolayısıyla süreklidir.

### Türev ve İntegral  

**Tanım (Türev).**  

$$
f'(a)=\lim_{h\to 0}\frac{f(a+h)-f(a)}{h}
$$  

**Tanım (İntegral).**  

$$
\int_a^b f(x)\,dx
$$  

**Kuantumdaki Rolü.**  
- Normalizasyon:  

$$
\int_{-\infty}^{\infty} |\psi(x,t)|^2 dx = 1
$$  

- Beklenen değer:  

$$
\langle A \rangle = \int_{-\infty}^{\infty} \psi^*(x)(A\psi)(x)\,dx
$$  

**Örnek**  

$$
\psi(x)=\mathcal{N}e^{-x^2/4a^2}, \quad (a>0)
$$  

Normalizasyon koşulu:  

$$
\mathcal{N}=(2\pi)^{-1/4} a^{-1/2}
$$  

**Örnek**  

$$
(\Delta x)^2 = \langle x^2\rangle - \langle x\rangle^2 = a^2
$$  

**Örnek**  

$$
\psi'(x)=-\frac{x}{2a^2}\psi(x), \quad
\psi''(x)=\left(\frac{x^2}{4a^4}-\frac{1}{2a^2}\right)\psi(x)
$$

### Fourier Dönüşümü  

**Tanım.**  

$$
\phi(p) = \frac{1}{\sqrt{2\pi\hbar}} \int_{-\infty}^{\infty} \psi(x) e^{-ipx/\hbar}\,dx
$$  

$$
\psi(x) = \frac{1}{\sqrt{2\pi\hbar}} \int_{-\infty}^{\infty} \phi(p) e^{ipx/\hbar}\,dp
$$  

**Özellikler.**  
- Doğrusallık:  

$$
\mathcal{F}\{af+bg\} = a\mathcal{F}\{f\} + b\mathcal{F}\{g\}
$$  

- Türev:  

$$
\mathcal{F}\{\tfrac{d}{dx}\psi\}(p)=\tfrac{ip}{\hbar}\phi(p)
$$  

- Parseval:  

$$
\int |\psi(x)|^2 dx = \int |\phi(p)|^2 dp
$$  

**Örnek (Gauss)**  

$$
\psi(x)=(2\pi)^{-1/4}a^{-1/2}e^{-x^2/4a^2}
$$  

Fourier dönüşümü yine Gauss’tur:  

$$
\Delta x = a, \quad \Delta p = \frac{\hbar}{2a}, \quad \Delta x\,\Delta p = \tfrac{\hbar}{2}
$$  

**Örnek (Düzlem Dalga)**  

$$
\psi_k(x)=\frac{1}{\sqrt{2\pi}}e^{ikx} 
\quad \Longleftrightarrow \quad 
\phi(p)=\delta(p-\hbar k)
$$  

---

### Bölüm Özeti  

- **Limit–Süreklilik:** Dalga fonksiyonlarının fiziksel kabulü için temel koşullar.  
- **Türev–İntegral:** Schrödinger denklemi, normalizasyon ve beklenen değer hesapları için zorunludur.  
- **Fourier Dönüşümü:** Konum ve momentum temsilleri arasında köprü kurar, belirsizlik ilkesini ortaya çıkarır.
