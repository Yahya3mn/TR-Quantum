## Olasılık ve İstatistik  

Kuantum mekaniği, deterministik (kesin) bir kuram değildir. Yani klasik mekaniğin aksine, bir parçacığın konumu, momentumu ya da spini ölçülmeden önce kesin olarak bilinemaz; yalnızca olasılıklarla ifade edilir. Bu nedenle, kuantumu anlamak için **olasılık** ve **istatistik** kavramlarını bilmek zorunludur.  

---

### 1. Olasılığın Temel Tanımı  

Olasılık, bir olayın gerçekleşme ihtimalini $0$ ile $1$ arasında bir sayı ile belirtir.  

- $0$: Olay imkânsızdır.  
- $1$: Olay kesin olarak gerçekleşir.  
- $0 < P < 1$: Olay belirli bir olasılıkla gerçekleşir.  

**Klasik örnekler:**  
- Bir paranın yazı gelme olasılığı:  

  $$
  P(\text{yazı}) = \tfrac{1}{2}
  $$  

- Bir zarın $6$ gelme olasılığı:  

  $$
  P(6) = \tfrac{1}{6}
  $$  

Bu tür olaylarda sonuçları kesin olarak bilemeyiz, ancak uzun vadede hangi sonuçların ne sıklıkta çıkacağını öngörebiliriz.  

---

### 2. Olasılık Dağılımları  

Bir rastgele değişkenin alabileceği değerler ve bu değerlere karşılık gelen olasılıkların tamamına **olasılık dağılımı** denir.  

- **Ayrık dağılım:**  
  Zar atıldığında $X = \{1,2,3,4,5,6\}$ değerlerini alır ve her biri için olasılık $\tfrac{1}{6}$’dır.  

  $$
  P(X=k) = \tfrac{1}{6}, \quad k=1,2,\dots,6
  $$  

- **Sürekli dağılım:**  
  Bir parçacığın konumu $x$ bir aralıkta herhangi bir değer alabilir. Olasılık yoğunluk fonksiyonu $f(x)$ ile verilir:  

  $$
  P(a \leq X \leq b) = \int_a^b f(x)\,dx, \quad \int_{-\infty}^{\infty} f(x)\,dx = 1
  $$  

**Kuantumda:** Dalga fonksiyonu $\psi(x)$, parçacığın konumunun olasılık genliğini ifade eder. Ölçüm olasılığı:  

$$
P(x) = |\psi(x)|^2
$$  

---

### 3. Ortalama Değer (Beklenen Değer)  

Beklenen değer, çok sayıda tekrar yapıldığında ortalama sonucu ifade eder.  

- **Zar örneği:**  

  $$
  \mathbb{E}[X] = \sum_{k=1}^6 k \cdot \tfrac{1}{6} = \tfrac{21}{6} = 3.5
  $$  

Yani zar atıldığında “ortalama” değer $3.5$’tir (hiçbir atışta çıkmamasına rağmen).  

- **Kuantumda:**  
  Bir gözlemlenebilir $A$ için beklenen değer:  

  $$
  \langle A \rangle = \langle \psi | A | \psi \rangle
  $$  

**Örnek:** Spin-$\tfrac{1}{2}$ parçacığı $|0\rangle$ durumunda iken $\sigma_z$ ölçülürse:  

$$
\langle \sigma_z \rangle = \langle 0 | \sigma_z | 0 \rangle = +1
$$  

---

### 4. Varyans ve Belirsizlik  

Varyans, ölçümlerin ortalama etrafında ne kadar “dağıldığını” gösterir.  

- **Zar için:**  

  Önce $\mathbb{E}[X^2]$ bulunur:  

  $$
  \mathbb{E}[X^2] = \tfrac{1^2+2^2+3^2+4^2+5^2+6^2}{6} = \tfrac{91}{6}
  $$  

  Daha sonra varyans:  

  $$
  \mathrm{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \tfrac{91}{6} - (3.5)^2 \approx 2.92
  $$  

- **Kuantumda:**  

  Bir gözlemlenebilir $A$ için varyans:  

  $$
  (\Delta A)^2 = \langle \psi | A^2 | \psi \rangle - \langle \psi | A | \psi \rangle^2
  $$  

Bu ifade, **Heisenberg belirsizlik ilkesinin** temelidir. Örneğin konum ($x$) ve momentum ($p$) ölçümlerinde:  

$$
\Delta x \, \Delta p \geq \tfrac{\hbar}{2}
$$  

---

### 5. Kuantum Ölçümlerinde Olasılık  

Kuantum ölçüm postülatına göre:  
- Bir sistem $|\psi\rangle$ durumunda olsun.  
- Bir gözlemlenebilir $A$’nın özvektörleri $|a_i\rangle$, özdeğerleri $a_i$ olsun.  
- Ölçüm yapıldığında, $a_i$ sonucunun elde edilme olasılığı:  

  $$
  P(a_i) = |\langle a_i | \psi \rangle|^2
  $$  

- Ölçümden sonra sistem, ilgili özvektör durumuna çöker:  

  $$
  |\psi\rangle \;\longrightarrow\; |a_i\rangle
  $$  

**Örnek:**  
Sistem $|\psi\rangle = \tfrac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ durumunda olsun.  

- Eğer $\sigma_z$ ölçülürse, sonuçlar:  
  - $+1$ değeri (durum $|0\rangle$) %50 olasılıkla,  
  - $-1$ değeri (durum $|1\rangle$) %50 olasılıkla elde edilir.  

---

### Genel Özet  

- Olasılık: Rastgele olayların gerçekleşme derecesi.  
- Ortalama değer: Çok sayıda tekrar sonunda beklenen ölçüm sonucu.  
- Varyans: Ölçüm sonuçlarının ortalama etrafında ne kadar yayıldığını gösterir.  
- Kuantumda ölçüm: Olasılık, iç çarpımın karesine göre belirlenir; sistem ölçümden sonra ilgili özvektöre çöker.  

Kuantum mekaniğinde olasılık ve istatistik yalnızca yardımcı kavramlar değil, kuramın temel taşlarıdır.
