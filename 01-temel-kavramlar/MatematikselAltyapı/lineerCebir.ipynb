## Giriş

Kuantum mekaniğini anlamak için matematik yalnızca bir araç değil, aynı zamanda vazgeçilmez bir dildir. Fiziksel olayları sözcüklerle tarif etmek çoğu zaman sezgisel kalır; ancak matematik, bu olayları kesin ve evrensel bir şekilde ifade etmemizi sağlar. Özellikle kuantum dünyasında, gündelik hayatta deneyimlemediğimiz olgularla karşılaştığımız için soyutlama gücüne ihtiyaç duyarız.

Bu yazıda amaç, karmaşık formüllerden çekinmeden kuantum mekaniğinin temelinde yatan matematiksel yapı taşlarını tanımaktır. Kavramlara adım adım yaklaşarak hem sezgisel bir bakış açısı kazanacak hem de ilerleyen bölümlerde daha derin sonuçları anlayabilmek için sağlam bir temel oluşturacağız.

---

## Lineer Cebir  

Kuantum mekaniğinde bir parçacığın durumu, gündelik hayatta kullandığımız sıradan sayılarla değil, **vektörler** adı verilen matematiksel nesnelerle ifade edilir. Ölçüm süreçleri ya da zaman içindeki evrim ise **matrisler** ve bunların tanımladığı lineer dönüşümler aracılığıyla modellenir. Dolayısıyla lineer cebir, kuantum mekaniğinin matematiksel dilini anlamanın temelini oluşturur.  

---

### 1. Vektörler  

Bir vektör, birden fazla bileşeni tek bir yapı altında toplayan matematiksel bir nesnedir.  

- Günlük hayatta: Bir ok olarak düşünülebilir; yönü ve büyüklüğü vardır.  
- Matematikte: Genellikle sütun biçiminde yazılır:  

$$
v = \begin{bmatrix}2 \\ 3\end{bmatrix}
$$

Bu, “x yönünde 2, y yönünde 3” anlamına gelir.  

Kuantum mekaniğinde parçacığın durumu da bir vektörle temsil edilir. Ancak burada bileşenler çoğu zaman **karmaşık sayılar**dır.  

Örnek:  

$$
\psi = \begin{bmatrix} \tfrac{1}{\sqrt{2}} \\ \tfrac{i}{\sqrt{2}} \end{bmatrix}
$$

---

### 2. İç Çarpım  

İki vektör arasındaki benzerliği ölçmek için **iç çarpım** tanımlanır:  

$$
\langle u , v \rangle = u_1 v_1 + u_2 v_2 + \dots
$$  

- İç çarpım, vektörlerin uzunluklarını ve aralarındaki açıyı belirler.  
- Kuantum mekaniğinde:  
  - İki durum arasındaki benzerliği ölçer.  
  - Ölçüm yapıldığında belli bir sonucu elde etme olasılığı iç çarpım üzerinden hesaplanır.  

Örnek:  

$$
u = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad v = \begin{bmatrix}0 \\ 1\end{bmatrix}
$$  

Bu durumda:  

$$
\langle u , v \rangle = 0
$$  

yani $u$ ve $v$ ortogonaldir. Kuantumda bu, birbirini dışlayan durumlara karşılık gelir.  

---

### 3. Matrisler  

Bir matris, satır ve sütunlardan oluşan bir tablo şeklinde yazılır:  

$$
A = \begin{bmatrix}1 & 2 \\ 3 & 4\end{bmatrix}
$$  

Matrisler, vektörler üzerinde işlem yaparak yeni vektörler üretir:  

$$
A \cdot v = v'
$$  

Kuantumda, matrisler fiziksel dönüşümleri (zaman evrimi, ölçüm, simetriler vb.) temsil eder.  

---

### 4. Lineer Dönüşümler  

Bir vektörü alıp başka bir vektöre dönüştüren kurallara **lineer dönüşüm** denir.  

- Örneğin: Her vektörü iki katına çıkarmak,  
- ya da: Vektörleri belirli bir açıyla döndürmek.  

Tüm bu dönüşümler matrislerle ifade edilir.  

---

### 5. Özvektörler ve Özdeğerler  

Bazı özel vektörler vardır ki, üzerlerine bir matris (operatör) uygulandığında yön değiştirmezler; yalnızca boyutları ölçeklenir:  

$$
A v = \lambda v
$$  

Burada:  
- $v$: **özvektör**  
- $\lambda$: **özdeğer**  

Kuantum mekaniğinde özvektörler, ölçüm yapıldığında sistemin bulunabileceği olası durumları; özdeğerler ise ölçümde elde edilen sayısal sonuçları temsil eder.  

Örneğin spin ölçümleri, özdeğer–özvektör ilişkisi ile modellenir.  

---

### 6. Tensor Çarpımı  

Tek parçacıklı sistemler vektörlerle temsil edilebilir. Ancak kuantum mekaniğinde genellikle birden fazla parçacıktan oluşan **bileşik sistemler** söz konusudur. Bu durumda sistemlerin durumları yalnızca toplama veya skaler çarpma ile birleştirilemez; bunun yerine **tensor çarpımı** kullanılır.  

#### Tanım  

Eğer  

$$
u = \begin{bmatrix} u_1 \\ u_2 \end{bmatrix}, \quad 
v = \begin{bmatrix} v_1 \\ v_2 \end{bmatrix}
$$  

ise, tensor çarpımı şu şekilde tanımlanır:  

$$
u \otimes v = 
\begin{bmatrix} 
u_1 v_1 \\ 
u_1 v_2 \\ 
u_2 v_1 \\ 
u_2 v_2 
\end{bmatrix}
$$  

Yani iki boyutlu iki vektörün birleşiminden dört boyutlu bir vektör elde edilir. Genel olarak, $m$ boyutlu bir vektör ile $n$ boyutlu bir vektörün tensor çarpımı $m \times n$ boyutlu bir vektör verir.  

---

#### Kuantum Mekaniğinde Kullanımı  

- Birinci parçacık $|0\rangle = \begin{bmatrix}1 \\ 0\end{bmatrix}$, ikinci parçacık $|1\rangle = \begin{bmatrix}0 \\ 1\end{bmatrix}$ durumunda ise:  

$$
|0\rangle \otimes |1\rangle = |01\rangle =
\begin{bmatrix}0 \\ 1 \\ 0 \\ 0\end{bmatrix}
$$  

- Tek parçacıklı bir sistemin boyutu $2$ ise, iki parçacıklı sistemin boyutu $2 \times 2 = 4$ olur. Genel olarak $n$ parçacıklı bir sistemin boyutu $2^n$ şeklindedir.  

---

#### Fiziksel Anlam  

Tensor çarpımı, çok parçacıklı sistemlerin birleşimini tanımlamanın matematiksel yoludur.  

- Sistemler bağımsızsa, birleşik sistemin durumu tensor çarpımı ile yazılır.  
- Sistemler etkileşimli veya **dolanık (entangled)** ise, yine tensor çarpımı kullanılır; fakat elde edilen vektör çoğu durumda tekil çarpımlara ayrıştırılamaz.  

---

### Genel Özet  

- **İç çarpım**: İki vektör → skaler. Benzerlik ve olasılık hesapları.  
- **Matrisler ve lineer dönüşümler**: Vektörler üzerinde işlem yapan yapılar. Fiziksel dönüşümler ve ölçümler.  
- **Özvektörler ve özdeğerler**: Ölçüm sonuçlarının matematiksel temsili.  
- **Tensor çarpımı**: İki vektör → daha yüksek boyutlu vektör. Çok parçacıklı sistemlerin matematiksel ifadesi.  

Bu nedenle kuantum mekaniğinde **iç çarpım** ölçüm olasılıklarıyla, **tensor çarpımı** ise bileşik sistemlerin yapısıyla ilişkilidir.
